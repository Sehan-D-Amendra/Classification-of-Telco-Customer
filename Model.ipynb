```python
yes_no_mapping = {'Yes': 1, 'No': 0}

Customer_Churn_data['Partner'] = Customer_Churn_data['Partner'].map(yes_no_mapping)
Customer_Churn_data['Dependents'] = Customer_Churn_data['Dependents'].map(yes_no_mapping)
Customer_Churn_data['PhoneService'] = Customer_Churn_data['PhoneService'].map(yes_no_mapping)
```

```python
InternetServiceMapping = {'Yes' : 1, 'No' : 0 , 'No internet service' : 0}

Customer_Churn_data['OnlineSecurity'] = Customer_Churn_data['OnlineSecurity'].map(InternetServiceMapping)
Customer_Churn_data['OnlineBackup'] = Customer_Churn_data['OnlineBackup'].map(InternetServiceMapping)
Customer_Churn_data['DeviceProtection'] = Customer_Churn_data['DeviceProtection'].map(InternetServiceMapping)
Customer_Churn_data['TechSupport'] = Customer_Churn_data['TechSupport'].map(InternetServiceMapping)
Customer_Churn_data['StreamingTV'] = Customer_Churn_data['StreamingTV'].map(InternetServiceMapping)
Customer_Churn_data['StreamingMovies'] = Customer_Churn_data['StreamingMovies'].map(InternetServiceMapping)
```

```python
PhoneServiceMapping = {'Yes' : 1, 'No' : 0 , 'No phone service' : 0}

Customer_Churn_data['MultipleLines'] = Customer_Churn_data['MultipleLines'].map(PhoneServiceMapping)
```

```python
Customer_Churn_data
```

```python
ContractMapping = {'Month-to-month' : 'MM', 'One year' : 'OY', 'Two year' : 'TY'}

Customer_Churn_data['Contract'] = Customer_Churn_data['Contract'].map(ContractMapping)
```

```python
Customer_Churn_data['PaperlessBilling'] = Customer_Churn_data['PaperlessBilling'].map(yes_no_mapping)
```

```python
Customer_Churn_data['PaymentMethod'].value_counts()
```

```python
PaymentMapping = {'Bank transfer (automatic)' : 'Bank', 'Credit card (automatic)' : 'Card', 'Electronic check' : 'Check', 'Mailed check' : 'Check'}

Customer_Churn_data['PaymentMethod'] = Customer_Churn_data['PaymentMethod'].map(PaymentMapping)
```

```python
Customer_Churn_data['Churn'] = Customer_Churn_data['Churn'].map(yes_no_mapping)
```

```python
Customer_Churn_data
```

```python
sex = pd.get_dummies(Customer_Churn_data['gender'], drop_first=True)
sex
```

```python
sex['Male'] = sex['Male'].map({True:1,False:0})
```

```python
sex
```

```python
contract = pd.get_dummies(Customer_Churn_data['Contract'], drop_first=True)
```

```python
contract
```

```python
contract['OY'] = contract['OY'].map({True:1,False:0})
contract['TY'] = contract['TY'].map({True:1,False:0})
```

```python
contract
```

```python
paymethod = pd.get_dummies(Customer_Churn_data['PaymentMethod'], drop_first=True)
```

```python
paymethod
```

```python
paymethod['Card'] = paymethod['Card'].map({True:1,False:0})
paymethod['Check'] = paymethod['Check'].map({True:1,False:0})
```

```python
paymethod
```

```python
Customer_Churn_data = pd.concat([Customer_Churn_data,sex,contract,paymethod], axis=1)
```

```python
Customer_Churn_data.head()
```

```python
Customer_Churn_data.drop(['customerID', 'gender', 'PaymentMethod', 'Contract'], axis = 1, inplace=True)
```

```python
Customer_Churn_data
```

```python
DSLFiberMapping = {'DSL' : 1, 'Fiber optic' : 1, 'No' : 0}

Customer_Churn_data['InternetService'] = Customer_Churn_data['InternetService'].map(DSLFiberMapping)
```

```python
Customer_Churn_data
```

# Encording

```python
from sklearn.preprocessing import LabelEncoder
lblEnc = LabelEncoder()
churn_data_tf = Customer_Churn_data.copy()

for i in Customer_Churn_data.columns:
    churn_data_tf[i] = lblEnc.fit_transform(Customer_Churn_data[i])
```

```python
churn_data_tf
```

```python
Customer_Churn_data
```

```python
X = churn_data_tf.drop('Churn', axis = 1)
Y = churn_data_tf['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 2)
```

```python
print(X_train.shape, '-----> Training Feature Shape')
print(X_test.shape, '-----> Testing Feature Shape')
print(Y_train.shape, '-----> Training Labels Shape')
print(Y_test.shape, '-----> Training Labels Shape')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
churn_log_model = LogisticRegression()
```

```python
churn_log_model.fit(X_train, Y_train)
```

```python
churn_predictions = churn_log_model.predict(X_test)
```

```python
churn_predictions
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, churn_predictions))
```

```python
import sklearn.metrics as sm
```

```python
accuracy_of_churn_model = sm.accuracy_score(Y_test, churn_predictions)

print('Accuracy Score ==', str(accuracy_of_churn_model))
```
